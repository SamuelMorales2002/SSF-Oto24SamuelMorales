```python
#6.1.1
import numpy as np

n_puntos = 10**6
radio = 1.0

x = np.random.random(n_puntos)
y = np.random.random(n_puntos)
z = np.random.random(n_puntos)

dentro_esfera = (x**2 + y**2 + z**2) <= radio**2

puntos_dentro = np.sum(dentro_esfera)

fraccion_dentro = puntos_dentro / n_puntos

vol_cubo = 1**3

vol_cuarto_esfera = fraccion_dentro * vol_cubo

vol_esfera = vol_cuarto_esfera * 4

print(f"Número total de puntos: {n_puntos}")
print(f"Puntos dentro del cuarto de esfera: {puntos_dentro}")
print(f"Volumen estimado del cuarto de esfera: {vol_cuarto_esfera:.5f}")
print(f"Volumen estimado de la esfera completa: {vol_esfera:.5f}")
```

```python
#6.1
import numpy as np
import matplotlib.pyplot as plt

n_puntos = 5000
a = 0
b = np.pi

x_aleatorios = np.random.uniform(a, b, n_puntos)
y_aleatorios = np.random.uniform(0, 1, n_puntos)
y_curva = np.sin(x_aleatorios)

dentro = y_aleatorios <= y_curva
fuera = y_aleatorios > y_curva

x_curva = np.linspace(a, b, 1000)
y_exacta = np.sin(x_curva)

plt.figure(figsize=(10, 6))

plt.plot(x_curva, y_exacta, color="blue", label="y = sin(x)")

plt.scatter(x_aleatorios[dentro], y_aleatorios[dentro], color="black", s=1, label="Puntos debajo de sin(x)")
plt.scatter(x_aleatorios[fuera], y_aleatorios[fuera], color="red", s=1, label="Puntos fuera de sin(x)")

plt.title("Método Monte Carlo para la Integral de sin(x)", fontsize=14)
plt.xlabel("x", fontsize=10)
plt.ylabel("y", fontsize=10)
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
